```python
import importlib
import RAGbuilder
importlib.reload(RAGbuilder)

from RAGbuilder import load_and_split_md

def debug_chunks(file_path):
    docs = load_and_split_md(file_path)
    print(f"📄 파일명: {file_path}")
    print(f"✂️ 총 생성된 청크 수: {len(docs)}개")
    print("-" * 50)

    for i, doc in enumerate(docs[:5]):  # 처음 5개만 확인
        content = doc.page_content.replace("\n", " ")
        print(f"[Chunk {i}]")
        print(f"📏 길이: {len(doc.page_content)}자")
        print(f"📝 내용: {content[:200]}...") # 앞부분 200자만 출력
        
        # overlap 확인을 위해 다음 청크 시작 부분 보여주기
        if i < len(docs) - 1:
            next_start = docs[i+1].page_content[:30].replace("\n", " ")
            print(f"🔗 다음 청크와의 연결부: ... [{next_start}]")
        
        print("-" * 50)
```

```python
if __name__ == "__main__":
    test_file = "data/(4-01) 취업규칙_260121.md"
    debug_chunks(test_file)
```
